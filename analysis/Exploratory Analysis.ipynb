```python
# Import dependencies
import pandas as pd
```

```python
# Load the data
data_to_load = "Data/meteorite_landings.csv"
meteorite_data = pd.read_csv(data_to_load)
meteorite_data.head()
```

```python
# Get row count
len(meteorite_data.index)
```

```python
# Sort by heaviest meteorite
heaviest = meteorite_data.sort_values(["mass (g)"], ascending=False)
heaviest.head()
```

```python
# Sort by lightest meteorite
lightest = meteorite_data.sort_values(["mass (g)"], ascending=True)
lightest.head()
```
